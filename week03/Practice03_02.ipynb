```python
from google.colab import drive  # 구글 드라이브 마운트
drive.mount("/content/drive", force_remount=True)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/080228-master/deeplearning/dataset/pima-indians-diabetes.csv',
                 names = ['pregnant', 'plasma', 'pressure', 'thickness', 'insulin', 'BMI', 'pedigree', 'age', 'class'])
age_mean = df[['age']].mean() # 평균 나이 계산
print('Average age:', round(age_mean.age, 2))
print()

low_age = df[df['age'] <= age_mean.age] # 평균 age 이하
high_age = df[df['age'] >= age_mean.age] # 평균 age 이상

# 선 그리기
low_age.thickness.plot.hist(bins=50, alpha=1.0)
high_age.thickness.plot.hist(bins=50, alpha=1.0)

# 데이터 종류 : low_age, high_age
plt.legend(['low_age', 'high_age'], loc='upper right')
plt.xlabel('thickness') # x축은 thickness
plt.show()
```
