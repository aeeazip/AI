```python
from google.colab import drive  # 구글 드라이브 마운트
drive.mount("/content/drive", force_remount=True)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/080228-master/deeplearning/dataset/pima-indians-diabetes.csv',
                 names = ['pregnant', 'plasma', 'pressure', 'thickness', 'insulin', 'BMI', 'pedigree', 'age', 'class'])
data1 = df[['pregnant', 'plasma']].values # pregnant, plasma만 추출

# data1에서 저장된 5개만 분리
data2 = data1[:5]
print(data2)
print(data2.shape)
print(type(data2))

# data2의 첫 번째 열들의 값 x1에 저장
x1 = list()
for i in range(data2.shape[0]):
  x1.append(data2[i][0])
print(x1)

# data2의 두 번째 열들의 값 x2에 저장
x2 = list()
for i in range(data2.shape[0]):
  x2.append(data2[i][1])
print(x2)
print('---------------------------------------')

plt.plot(x1, x2, "o") # 점 찍기
plt.plot(x1, x2, color='blue')  # 선 그리기
plt.show()
```
