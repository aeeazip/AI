```python
import requests
import pandas as pd
import os
from bs4 import BeautifulSoup
from google.colab import drive
drive.mount("/content/drive", force_remount=True)

# 영역별 크롤링 시도
def crawling(sid1, label, df):
  url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=" + str(sid1)

  # 크롤링 자동 방지 해제
  headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"
  }

  result = requests.get(url, headers=headers)
  soup = BeautifulSoup(result.text, "html.parser")
  print(soup)

  news_titles = soup.find_all(class_="sa_text_strong")
  # print(news_titles)
  # print("총 뉴스 제목 개수: ", len(news_titles))

  for item in news_titles:
    texts = item.get_text()
    texts = texts.replace(",", "")
    texts = texts.replace("...", " ")
    texts = texts.replace("`'", "")
    # print(texts + ", label : " + str(label))

    onedata = {"title" : texts, "label": str(label)}
    onedata_df = pd.DataFrame([onedata]) # 딕셔너리를 데이터프레임으로 변환
    df = pd.concat([df, onedata_df])

  return df

# 메인
csv_path = "/content/drive/MyDrive/test_SelectionDataset.csv"

if os.path.isfile(csv_path):
    df = pd.read_csv(csv_path)
else:
    df = pd.DataFrame(columns=["title", "label"])

for i in range(100, 104):
  # 정치(100), 경제(101), 사회(102), 생활/문화(103)
  df = crawling(i, i - 100, df)
  df.to_csv(csv_path, encoding="UTF-8", index=False, header=True)

print("Done")
```
