```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 실습 과제1 : 시그모이드 함수
# data = [[2.0, 0.0], [4.0, 0.0], [6.0, 0.0], [8.0, 1.0], [10.0, 1.0], [12.0, 1.0], [14.0, 1.0]]

# x = [i[0] for i in data]
# y = [i[1] for i in data]

# x_data = np.array(x)
# y_data = np.array(y)

# a = 0
# b = 0
# lr = 0.05

# def error_sum():
#   return ((sigmoid(a * x_data + b) - y_data) ** 2).mean()

# def sigmoid(x):
#   return 1 / (1 + np.e ** (-x))

# epochs = 10001
# loss = []
# for i in range(epochs):
#   a_diff = 1 / len(x_data) * sum(x_data * (sigmoid(a * x_data + b) - y_data))
#   b_diff = 1 / len(x_data) * sum(sigmoid(a * x_data + b) - y_data)

#   loss.append(error_sum())

#   a = a - lr * a_diff
#   b = b - lr * b_diff

#   if i % 1000 == 0:
#     print("epoch=%d, 기울기=%.4f, 절편=%.4f" % (i, a, b))
#     print("Error value: %.4f" % (loss[i]))

# plt.plot(np.arange(0, 10001), np.array(loss))
# plt.xlabel("epoch (Loop Num)")
# plt.ylabel("loss[]")
# plt.ylim(0.0, 0.2)
# plt.show()

# for i in range(len(x)):
#   print("Input: %f, Real: %f, Expect: %.4f" % (x[i], y[i], sigmoid(a * x[i] + b)))


# 실습 과제2
data = [[2.0, 0.0], [4.0, 0.0], [6.0, 0.0], [8.0, 1.0], [10.0, 1.0], [12.0, 1.0], [14.0, 0.0]]

x = [i[0] for i in data]
y = [i[1] for i in data]

x_data = np.array(x)
y_data = np.array(y)

a = 0
b = 0
lr = 0.05

def error_sum():
  return ((sigmoid(a * x_data + b) - y_data) ** 2).mean()

def sigmoid(x):
  return 1 / (1 + np.e ** (-x))

epochs = 10001
loss = []
for i in range(epochs):
  a_diff = 1 / len(x_data) * sum(x_data * (sigmoid(a * x_data + b) - y_data))
  b_diff = 1 / len(x_data) * sum(sigmoid(a * x_data + b) - y_data)

  loss.append(error_sum())

  if i % 1000 == 0:
    print("epoch=%d, 기울기=%.4f, 절편=%.4f" % (i, a, b))
    print("Error value: %.4f" % (loss[i]))

  if loss[i] > loss[i - 1]:
    lr -= 0.01

  a = a - lr * a_diff
  b = b - lr * b_diff
# 실제 데이터, 예측 데이터를 활용한 그래프
plt.scatter(x, y)
x_range = np.arange(0, 15)
plt.plot(x_range, np.array([sigmoid(a * x + b) for x in x_range]))
plt.show()

# loss 그래프
plt.plot(np.arange(0, 10001), np.array(loss))
plt.xlabel("epoch (Loop Num)")
plt.ylabel("loss[]")
plt.ylim(0.0, 0.25)
plt.show()

for i in range(len(x)):
  print("Input: %f, Real: %f, Expect: %.4f" % (x[i], y[i], sigmoid(a * x[i] + b)))
```
